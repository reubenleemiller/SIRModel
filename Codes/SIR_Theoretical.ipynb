```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S = np.zeros(101)
I = np.zeros(101)
R = np.zeros(101)
time = np.arange(0, 101)

S[0] = 10000
I[0] = 10
R[0] = 0

beta = 0.00003
alpha = 0.05
delta = 0.001

for i in range(100):
    S[i+1] = S[i] - beta * I[i] * S[i] + delta * R[i]
    I[i+1] = I[i] + beta * I[i] * S[i] - alpha * I[i]
    R[i+1] = R[i] + alpha * I[i] - delta * R[i]

plt.figure(1)
plt.plot(time, S, 'b',linewidth=0.4, label='Susceptible')
plt.plot(time, I, 'm',linewidth=0.4, label='Infected')
plt.plot(time, R, 'r',linewidth=0.4, label='Recovered')
plt.title('Plot of # of (S, I, R) People Against Time', fontname='serif')
plt.xlabel('Time (days)', fontname='serif')
plt.ylabel('S, I, R (# of people)', fontname='serif')
plt.xticks(np.arange(0, 101, 10))
plt.yticks(np.arange(0, 10001, 1000))
plt.legend()

plt.show()
```
