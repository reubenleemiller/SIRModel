```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
S = np.zeros(101)
I = np.zeros(101)
R = np.zeros(101)
time = np.arange(0, 101)

csv_data = pd.read_csv("SARS_Data.csv")
Days = csv_data["Days"]
T = csv_data["T"]
F = csv_data["F"]
H = csv_data["H"]
O = csv_data["O"]
I_1 = T + F + H + O

S[0] = 251
I[0] = 1
R[0] = 0

beta = 0.0004
gamma = 0.002

for i in range(100):
    S[i+1] = S[i] - beta * I[i] * S[i] 
    I[i+1] = I[i] + beta * I[i] * S[i] - gamma * I[i]
    R[i+1] = R[i] + gamma * I[i] 

plt.figure(1)
plt.plot(Days,I_1,label="Experimental",linestyle="",linewidth=0.2,marker="o",markersize=2)
plt.plot(time, I, 'm',linewidth=0.4, label='Theoretical')
plt.title('Plot of # of Infected People Against Time', fontname='serif')
plt.xlabel('t (days)', fontname='serif')
plt.ylabel('# of Infected People', fontname='serif')
plt.legend()
plt.grid(True,linewidth=0.2)
plt.show()
```
