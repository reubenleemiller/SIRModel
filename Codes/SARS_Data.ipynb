```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
csv_data = pd.read_csv("SARS_Data.csv")
Days = csv_data["Days"]
T = csv_data["T"]
F = csv_data["F"]
H = csv_data["H"]
O = csv_data["O"]
coefficients = np.polyfit(Days, H, deg=1)
polynomial_function = np.poly1d(coefficients)
y_poly_pred = polynomial_function(Days)
plt.figure()
plt.title("Plot of New Infections From Health Care Setting (H)",family='serif')
plt.xlabel("Time (Days)",family='serif')
plt.ylabel("New Cases",family='serif')
plt.plot(Days,H,label="H",linestyle="",linewidth=0.2,marker="o",markersize=2)
plt.plot(Days, y_poly_pred, label="Linear fit for H", linestyle="-",linewidth=0.5, color="red")
plt.legend(loc='best')
plt.xticks(family='serif')
plt.yticks(family='serif')
equation_text = f"y = {coefficients[0]:.4f}x + {coefficients[1]:.4f}"
plt.show()
print(equation_text)
```
